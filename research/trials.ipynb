```python
def get_predict(x:int, y:int) -> int:
    return x*y
```

```python
get_predict(2,2)
```

```python
get_predict(2,'2') 
```

ensure_annotations

```python
from ensure import ensure_annotations
@ensure_annotations
def get_predict(x:int, y:int) -> int:
    return x*y
```

```python
get_predict(2,'2')
```
